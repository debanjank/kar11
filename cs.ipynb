## Removing outliers

```python
def remove_outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    return df_out
```

## Correlation for numerical columns for target 1.

```python
num_df = df_1[['AMT_INCOME_TOTAL', 'AMT_GOODS_PRICE', 'AMT_CREDIT','AMT_ANNUITY']]
corr_matrix = num_df.corr()
sns.heatmap(corr_matrix, annot =True)
```

```python
corrdf = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
corrdf = corrdf.unstack().reset_index()
corrdf.columns = [ 'VAR1', 'VAR2','CORRELATION']
corrdf
```

```python
corrdf.dropna(subset = ['CORRELATION'], inplace = True)
```

```python
corrdf.Correlation = abs(corrdf.CORRELATION)
corrdf.sort_values('CORRELATION', ascending = False)
corrdf.CORRELATION = abs(corrdf.CORRELATION)
corrdf.sort_values('CORRELATION', ascending = False)
```
